## cleaning and importing the data

```python
import pandas as pd
import tensorflow as tf
```

```python
diabetes_db=pd.read_csv("pima-indians-diabetes.csv")
```

```python
diabetes_db.columns
```

```python
columns_to_normal=['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
diabetes_db[columns_to_normal]=diabetes_db[columns_to_normal].apply(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
num_preg=tf.feature_column.numeric_column('Number_pregnant')
```

```python
Glucose_concentration=tf.feature_column.numeric_column('Glucose_concentration')
```

```python
Blood_press=tf.feature_column.numeric_column('Blood_pressure')
```

```python
Triceps=tf.feature_column.numeric_column('Triceps')
```

```python
Insulin = tf.feature_column.numeric_column('Insulin')
BMI = tf.feature_column.numeric_column('BMI')
Diabetes_pedigree = tf.feature_column.numeric_column('Pedigree')
Age = tf.feature_column.numeric_column('Age')
```

```python
Group=tf.feature_column.categorical_column_with_vocabulary_list('Group',['A','B','C','D'])
```

```python
feat_cols = [num_preg,Glucose_concentration,Blood_press,Triceps,Insulin,BMI,Diabetes_pedigree,Group,Age]
```

```python
from sklearn.model_selection import train_test_split
```

```python
MyData = diabetes_db.drop('Class',axis=1)
```

```python
Jiren=diabetes_db['Class']
```

```python
x_train,x_eval,y_train,y_eval=train_test_split(MyData,Jiren,test_size=0.3)
```

# Create,Train then evaluate the bot

```python
estimator=tf.estimator.LinearClassifier(feat_cols,n_classes=2)
```

```python
input_fn=tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,num_epochs=1000,shuffle=False)
```

```python
eval_fn=tf.estimator.inputs.pandas_input_fn(x=x_eval,y=y_eval,num_epochs=1000,shuffle=False)
```

```python
estimator.train(input_fn,steps=10000000)
```

```python
result=estimator.evaluate(input_fn=eval_fn)
```

```python
result
```

# Test the bot

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(
      x=x_eval,
      batch_size=10,
      num_epochs=1,
      shuffle=False)
```

```python
predictions = estimator.predict(pred_input_func)
```
